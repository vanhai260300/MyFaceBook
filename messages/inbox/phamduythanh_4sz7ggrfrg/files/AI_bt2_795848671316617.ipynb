```python
### Tạo 10 điểm dữ liệu ngẫu nhiên và vẽ chúng lên trục toạ độ. Tính w0, w1 và R-squared dùng thư viện Sklearn
```

```python


import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

x = np.random.randint(1,20, size=10)
x.sort()
n = np.size(x)

y = np.random.randint(1,20, size=n)
y.sort()

colors = np.random.rand(n)
area = 50
reg = LinearRegression().fit(x.reshape(-1,1),y)
y_pred = reg.intercept_ + reg.coef_[0]*x
r2Score = r2_score(y, y_pred)

print('w1 using Sklearn library is ',reg.coef_[0])
print('w0 using Sklearn library is ',reg.intercept_)
print('R2 score: ',r2Score)

plt.scatter(x, y, area, colors, marker="o")

plt.plot(x, y_pred, color="b")

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
### Tính w0 và w1 với công thức ở slide số 9 và in ra.
```

```python
def w0_w1(n,x,y):
    w1 = (n*sum(x*y) - x.sum()*y.sum())/ (n*sum(x*x) - x.sum()*x.sum())
    w0 = (y.sum() - w1*x.sum())/ (n)
    return [w0,w1]

w0, w1 = w0_w1(n,x,y) 
print ('w1 using fomular is: ',w1)
print ('w0 using fomular is: ',w0)
    
```

```python
###Vẽ đường thẳng với w0 và w1
```

```python
plt.scatter(x, y, area, colors, marker="o")

y_new = w0 + w1*x

plt.plot(x, y_new, color="b")

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
