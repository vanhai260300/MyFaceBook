```python
#Tạo 300 điểm dữ liệu với 4 phân cụm, dùng thư viện sklearn để tìm điểm centroid của 4 phân cụng đó và vẽ lên trục toạ độ.
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
points, y_true = make_blobs(n_samples=300, centers=4,cluster_std=1.5, random_state=1)
kmeans = KMeans(n_clusters=4)
kmeans.fit(points)
y_colors = kmeans.predict(points)
plt.scatter(points[:, 0], points[:, 1], c=y_colors, s=40, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=40, alpha=0.8)
plt.show()
print('shape input data :',points.shape)
```

```python
#Viết hàm dự đoán nếu có input là 1 điểm dữ liệu thì xác định nó thuộc centroid nào.
#Vẽ tất cả các điểm và các centroid lên trục toạ độ.
import numpy as np
from sklearn.metrics import pairwise_distances_argmin
x = [[5,-2]]
points = np.vstack([points, x])
plt.scatter(points[:,0],points[:,1], 20)
kmeans = KMeans(n_clusters=4, random_state=0).fit(points)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], color = 'r')
subList = x[0]
plt.scatter(subList[0], subList[1], c='yellow', s=75, alpha=0.5);
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5);
print(centers)
labels = pairwise_distances_argmin(x, centers)
print("diem ",x, " gan centroid",centers[labels]," nhat")
```
