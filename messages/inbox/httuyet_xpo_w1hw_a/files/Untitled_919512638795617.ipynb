```python
#cau1
import pandas as pd
X = pd.read_csv('income_spend.csv')
X.head(10)
import matplotlib.pyplot as plt
plt.scatter(X['INCOME'], X['SPEND'], s=50);
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=1)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X['INCOME'], X['SPEND'], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)
```

```python
#cau 2
import numpy as np
def timdiemgannhat(X, centers):
    cum1 = list(range(X.shape[0]))
    cum1 = np.sqrt((X['INCOME'] - centers[0,0])**2+( X['SPEND'] - centers[0,1])**2)
    cum1 = sum(cum1)/(X.shape[0])
    return cum1
khoangcach = timdiemgannhat(X,centers)
print(khoangcach)
```
